```python
# Lösning uppgift 1a
# ------------------
# Funktion som testar vilka kombinationer som uppfyller pythagoras sats och returnerar en lista med dessa 
def uppfyller_pythagoras_sats(N):
    pythagoras_list = []
    # Kontrollerar alla mojliga kombinationer av abc
    for a in range(1, N + 1):
        for b in range(a, N + 1):
            for c in range(b, N + 1):
                # Kontrollerar om kombination abc uppfyller pythagoras sats
                if c**2 == a**2 + b**2:
                    pythagoras_list.append([a, b, c]) #  Lagg till kombination till lista
    
    return pythagoras_list
# Tar user input och kallar funktionen med numret som argument
N = int(input("Ange talet N: "))
kombinationer = uppfyller_pythagoras_sats(N)

# Kontrollerar om giltiga kombinationer har hittats och printar dessa
if kombinationer:
    print(f"Det finns {len(kombinationer)} heltalskombinationer som uppfyller Pythagoras sats:")
    for i, kombinationer in enumerate(kombinationer, start=1):
        print(f"{i}. {kombinationer}")
else:
    print("Inga heltalskombinationer hittades som uppfyller Pythagoras sats för det givna intervallet.")
```

```python
# Lösning uppgift 1b
# ------------------
def omvandla(tal):
    # Dictionary som innehaller hexadecimala varden och motsv binara varden som key-value pairs
    hexadecimal_to_binary = {
        '0': '0000', '1': '0001', '2': '0010', '3': '0011',
        '4': '0100', '5': '0101', '6': '0110', '7': '0111',
        '8': '1000', '9': '1001', 'A': '1010', 'B': '1011',
        'C': '1100', 'D': '1101', 'E': '1110', 'F': '1111'
    }
    binary_result = ''  # Initialiserar med tom string
    for char in tal:
        # Om char finns i dictionary, addera motsv binart varde
        if char.upper() in hexadecimal_to_binary:
            binary_result += hexadecimal_to_binary[char.upper()]
        else:
            pass # Om ogiltigt hexadecimaltal, ignorera
    return binary_result

# Huvudprogram anropar funktionen omvandla() med user input
hexadecimal_input = input("Ange ett hexadecimalt tal som ska omvandlas: ")
binary_output = omvandla(hexadecimal_input)
print(f"Det binära talet blir {binary_output}")
```

```python
# Lösning uppgift 2a
# ------------------
import matplotlib.pyplot as plt
import random  # for funktionen random.uniform()

# Funktion for att hitta minsta varde
def find_min_value(list):
    if not list:  # Kolla att listan inte ar tom
        return None
    min_value = list[0]  # Initialisera min value till forsta element i listan
    for number in list:
        if number < min_value:
            min_value = number  # Uppdatera min value om mindre hittas

    return min_value

def test_uniform(min_value, max_value, simulations):
    A_vinn = 0
    B_vinn = 0
    C_vinn = 0
    A_vinn_percent = []  # Lista for att lagra andelen A-vinster

    for i in range(simulations):
        D = random.uniform(min_value, max_value)  # Slumpmassigt tal inom det givna intervallet
        A = random.uniform(min_value, max_value)  # Skapa slumpmassiga tal för ABC
        B = random.uniform(min_value, max_value)
        C = random.uniform(min_value, max_value)

        # Hitta vilket av talen A, B och C som ligger narmast D
        closest = find_min_value([abs(A - D), abs(B - D), abs(C - D)])
        # Inkrementera talet narmast D
        if closest == abs(A - D):
            A_vinn += 1
        elif closest == abs(B - D):
            B_vinn += 1
        else:
            C_vinn += 1
        
        A_vinn_percent.append(A_vinn / (i + 1))  # Berakna andelen A-vinster och spara i listan

    print("\nResultat:")
    print("-----------")
    print(f"A_vinn = {A_vinn}, B_vinn = {B_vinn}, C_vinn = {C_vinn}")

    # Plotta andelen A-vinster som funktion av antalet simuleringar
    plt.plot(range(1, simulations + 1), A_vinn_percent, label='Andel A_vinn')
    plt.xlabel('antal simuleringar')
    plt.ylabel('Andel A_vinn som funktion av antalet simuleringar')
    plt.legend()
    plt.grid(True)
    plt.show()

# User input min, max och antal simuleringar
min_input = float(input("Ange min: "))
max_input = float(input("Ange max: "))
simulations_input = int(input("Ange antalet simuleringar N: "))

# Anropa funktionen m user input
test_uniform(min_input, max_input, simulations_input)
```

```python
# Lösning uppgift 3a
# ------------------
import csv

# Tar ett filnamn som argument och returnerar en lista med innehållet
def read_file(file_name):
    data = []
    with open(file_name, 'r', newline='') as file:
        reader = csv.reader(file, delimiter=';')
        for row in reader:
            data.append(row)
    return data				

kpi = 'kpi.csv'	
kpiData = read_file(kpi)

print("Innehåll i kpi:")
for i in range(3):
    print(kpiData[i])  # Skriver ut 3 forsta raderna
```

```python
# Lösning uppgift 3b
# ------------------


# Funktion for att hitta minsta varde
def min_value(list):
    if not list:  # Kolla att listan inte ar tom
        return None
    min_value = list[0]  # Initialisera min value till forsta element i listan
    for number in list:
        if number < min_value:
            min_value = number  # Uppdatera min value om mindre hittas

    return min_value

# Funktion for att hitta maxvarde
def max_value(list):
    if not list: 
        return None
    max_value = list[0] 
    for number in list:
        if number > max_value:
            max_value = number  # Uppdatera min value om storre hittas

    return max_value	

# Funktion for att hitta medel
def medel(list):
    if not list:
        return 0
    total = sum_function(list)
    count = len(list)
    return total / count

# Sorteringsfunktion
def sort_function(list):                                            
    n = len(list)
    for i in range(n):
        for j in range(1, n):
            if list[j - 1] > list[j]:
                list[j], list[j - 1] = list[j - 1], list[j]
    return list

# Summerar
def sum_function(lst):
    total = 0
    for number in lst:
        total += number
    return total

# Funktion för att berakna arlig inflation
def calculate_yearly_inflation(structured_data):
    yearly_inflation = {}
    sorted_years = sort_function([year for year in structured_data.keys() if year != 2023])
    
    for i in range(len(sorted_years) - 1):
        current_year = sorted_years[i]
        next_year = sorted_years[i + 1]
        
        # Berakna medel-KPI för nuvarande och nastkommande ar
        mean_kpi_current = sum_function(list(structured_data[current_year].values())) / len(structured_data[current_year])
        mean_kpi_next = sum_function(list(structured_data[next_year].values())) / len(structured_data[next_year])
        
        # Berakna arlig inflation
        mean_inflation = ((mean_kpi_next - mean_kpi_current) / mean_kpi_current) * 100
        
        yearly_inflation[current_year] = mean_inflation
    
    return yearly_inflation

# Funktion for att berakna manatlig inflation
def calculate_monthly_inflation(structured_data):
    monthly_inflation = {}
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'Maj', 'Jun', 'Jul', 'Aug', 'Sep', 'Okt', 'Nov', 'Dec']

    sorted_years = sort_function([year for year in structured_data.keys() if year != 2023])

    for year in sorted_years:
        months_data = list(structured_data[year].keys())
        months_data.sort(key=lambda x: months.index(x) if x in months else -1)

        # Fyll pa manader som saknas med vardet 0.0
        for month in months:
            if month not in months_data:
                structured_data[year][month] = 0.0

        months_data = list(structured_data[year].keys())
        months_data.sort(key=lambda x: months.index(x))

        for i in range(1, len(months_data) - 1):
            current_month = months_data[i]
            next_month = months_data[i + 1]

            kpi_current = structured_data[year][current_month]
            kpi_next = structured_data[year][next_month]

            # Berakna manatlig inflation
            month_inflation = ((kpi_next - kpi_current) / kpi_current) * 100
            monthly_inflation[f"{year}/{current_month}"] = month_inflation
    
    return monthly_inflation

# Funktion for att hitta ar med hogst och lagst KPI
def find_extreme_kpi_years(structured_data):
    kpi_values = {year: sum_function(list(values.values())) for year, values in structured_data.items()}
    max_kpi_year = max_value(list(kpi_values.keys()))
    min_kpi_year = min_value(list(kpi_values.keys()))
    return max_kpi_year, min_kpi_year

# Funktion for att hitta manad med hogst och lagst KPI
def find_extreme_inflation_months(monthly_inflation):
    max_inflation_month = max_value(list(monthly_inflation.keys()))
    min_inflation_month = min_value(list(monthly_inflation.keys()))
    return max_inflation_month, min_inflation_month

# Beraknar medelinflation for perioden
def calculate_mean_inflation_yearly(yearly_inflation):
    relevant_years = [year for year in yearly_inflation.keys() if year <= 2023]
    inflation_values = [yearly_inflation[year] for year in relevant_years]
    mean_inflation = medel(inflation_values)
    return mean_inflation

# Strukturerar data
kpiData = read_file(kpi)
structured_data = {}
header = kpiData[0]
kpiData = kpiData[1:]

for row in kpiData:
    year = int(row[0])
    structured_data[year] = {}
    
    for i, month in enumerate(header[1:], start=1):
        if row[i] != '':
            try:
                structured_data[year][month] = float(row[i])
            except ValueError:
                structured_data[year][month] = 0.0
        else:
            structured_data[year][month] = 0.0

# Berakna inflation
yearly_inflation = calculate_yearly_inflation(structured_data)
monthly_inflation = calculate_monthly_inflation(structured_data)
mean_inflation_yearly = calculate_mean_inflation_yearly(yearly_inflation)

# Hitta ar med hogsta och lagsta KPI
max_kpi_year, min_kpi_year = find_extreme_kpi_years(structured_data)

# Hitta manad med hogsta och lagsta inflation
max_inflation_month, min_inflation_month = find_extreme_inflation_months(monthly_inflation)

# Skriv ut resultat i tabell
print("-----------------------------------------------------------------------------------------")
print("Inflationen i Sverige under tidsperioden 1 januari 1980 -- september 2023 uttryckt i %")
print("-----------------------------------------------------------------------------------------")
print("\t\t\t\tårlig\t\t\t\tMånad\t\t\t\t")
print("\t\t\tHögst  år  Lägst  år\t\t\tHögst\tår / mån    Lägst år / mån\t")
print("-----------------------------------------------------------------------------------------")
print(f"Inflationstakt årlig   : {max_value(list(yearly_inflation.values())):.1f}  {max_kpi_year}  {min_value(list(yearly_inflation.values())):.1f}  {min_kpi_year}\t\t\t\t")
print(f"Inflationstakt månatlig: \t\t\t\t\t{max_value(list(monthly_inflation.values())):.1f}\t{max_inflation_month}   {min_value(list(monthly_inflation.values())):.1f}  {min_inflation_month}\t\t\t\t")
print(f"\nMedeldinflation under perioden (%): {mean_inflation_yearly:.1f}%")
```
